```python
#importing libraries

import pandas as pd #import the pandas library for data manipulation and analysis
import matplotlib.pyplot as plt #import the matplotlib library for data visualization
```

```python
#loading individual datasets

orders = pd.read_csv("orders.csv")
order_details = pd.read_csv("order_details.csv")
pizzas = pd.read_csv("pizzas.csv")
pizza_types = pd.read_csv("pizza_types.csv")
```

```python
#joining the four CSV files into a single data frame

df = order_details.merge(orders, on="order_id").merge(pizzas, on="pizza_id").merge(pizza_types, on="pizza_type_id")
```

```python
df #to display DataFrame content
```

```python
#convert date and time to datetime format

df["order_time"] = pd.to_datetime(df["date"] + " " + df["time"])
```

## Recommended Analysis For Pizza Place Sales Task

```python
#What is the total revenue/sales?

total_revenue = (df["price"] * df["quantity"]).sum() #multiply price by quantity and sum the results
print(f"1. Total Revenue: ${total_revenue:,.2f}") #print the total revenue in dollars
```

```python
#Find the total quantity sold.

total_quantity = df["quantity"].sum() #sum the quantity column
print(f"2. Total Quantity Sold: {total_quantity}") #print the total quantity sold
```

```python
#Find the total orders

total_orders = df["order_id"].nunique() #get the unique number of order_id
print(f"3. Total Orders: {total_orders}") #print the total number of orders
```

```python
#How many pizza types do they sell?

num_pizza_types = df["pizza_type_id"].nunique() #get the unique number of pizza_type_id
print(f"4. Number of Pizza Types: {num_pizza_types}") #print the number of pizza types
```

```python
#Find the average price of the pizzas.

average_price = total_revenue / total_quantity #divide total revenue by total quantity
print(f"5. Average Price: ${average_price:,.2f}") #print the average price in dollars
```

```python
#What are the peak hours of sales?

df["hour"] = df["order_time"].dt.hour #extract the hour from the order_time column
peak_hour = df["hour"].value_counts().idxmax() #get the hour with the most sales
print(f"6. Peak Hour of Sales: {peak_hour}:00") #print the peak hour of sales
```

```python
#Find the total sales made on each day of the week. Which day of the week is when sales are made the most?


df["day_of_week"] = df["order_time"].dt.day_name() #extract the day of the week from the order_time column
daily_sales = df.groupby("day_of_week")["quantity"].sum() #group by day of the week and sum the quantity to get daily sales
peak_day = daily_sales.idxmax() #get the day with the most sales
print(f"7. Peak Day of Sales: {peak_day} (Total Sales: {daily_sales.max()})") #print the peak day of sales along with the total sales
```

```python
#Find the top 5 bestselling pizzas.

top_pizzas = df.groupby("name")["quantity"] #group by pizza name according to quantity and pass into function top_pizzas
top_5_pizzas = top_pizzas.sum().sort_values(ascending=False).head(5) #sum the quantity to get total sales for each pizza
print("8. Top 5 Bestselling Pizzas:") #print the top 5 bestselling pizzas
print(top_5_pizzas)
```

```python
#Find the sales made in each month. Any trend noticeable?

df["month"] = df["order_time"].dt.to_period("M") #extract the month from the order_time column
monthly_sales = df.groupby("month")["quantity"].sum() #group by month and sum the quantity to get monthly sales
print("9. Sales Trend by Month:") #print the sales trend by month
print(monthly_sales)
```

```python
#Are there pizza types that are not doing well on the menu?

total_quantity_threshold = total_quantity * 0.01 #1% threshold #calculate the threshold for low-performing pizzas(1% of total quantity)
low_performing = df.groupby("name")["quantity"].sum() #group by pizza name and sum the quantity to get total sales for each pizza
low_performing = low_performing[low_performing < total_quantity_threshold].index.tolist() #filter pizzas with sales less than the threshold
print(f"10. Pizza Types Not Doing Well: {low_performing}") #print the pizza types not doing well
```
